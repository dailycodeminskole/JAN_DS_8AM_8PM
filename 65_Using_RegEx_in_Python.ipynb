# Group

```python
sentence = "Vikas works in CTS and earns Rs. 100000"
pattern = "(^[A-Z]\w{3,}).+(\d{6}$)"
ans = re.match(pattern,sentence)
print(ans)
print(ans.group(1))
print(ans.group(2))
```
