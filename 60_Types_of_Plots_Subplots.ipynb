# SUBPLOTS

```python
x = np.arange(1,10,.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,10,.1)
y = np.tan(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,10,.1)
y1 = np.tan(x)
y2 = np.sin(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# using subplots
```

```python
x = np.arange(1,10,.1)
y1 = np.tan(x)
y2 = np.sin(x)
y3 = np.cos(x)


fig , ax  = plt.subplots(1,3)

ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
```

```python
x = np.arange(1,10,.1)
y1 = np.tan(x)
y2 = np.sin(x)
y3 = np.cos(x)

plt.figure()
plt.subplot(3,1,1)
plt.plot(x,y1)


plt.subplot(3,1,2)
plt.plot(x,y2)
```

```python
x = np.arange(1,10,.1)
y1 = np.tan(x)
y2 = np.sin(x)
y3 = np.cos(x)


plt.figure()
plt.subplot(2,3,1)
plt.plot(x,y1)


plt.subplot(2,3,3)
plt.plot(x,y2)


plt.subplot(2,3,5)
plt.plot(x,y3)
```

```python
a = np.arange(-1,1,0.005)
b = a

a,b = np.meshgrid(a,b)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a,b,a**2+b**2)
plt.show()
```
