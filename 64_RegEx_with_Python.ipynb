# Extracting the phone numbers from the String

```python
# 615-555-7164
pattern =  "\d{3}-\d{3}-\d{4}"
print(re.match(pattern,data)) #only at the beginning of the string
```

```python
# 615-555-7164
pattern =  "\d{3}-\d{3}-\d{4}"
print(re.search(pattern,data)) #1st pattern in the string
```

```python
# 615-555-7164
pattern =  "\d{3}-\d{3}-\d{4}"
print(re.findall(pattern,data)) #1st pattern in the string
```

```python
# 615-555-7164
pattern =  "\d{3}-\d{3}-\d{4}"
ans  = re.finditer(pattern,data) # this will form a list
for i in ans:
    print(i)
```

```python
# 615-555-7164
pattern =  "\d{3}-\d{3}-\d{4}"
ans  = re.finditer(pattern,data) # this will form a list
for i in enumerate(ans):
    print(i)
```

```python
# 615-555-7164
pattern =  "\d{3}-\d{3}-\d{4}"
ans  = re.finditer(pattern,data) # this will form a list
for i,num in enumerate(ans):
    print(i,num.start(),num.end(),num.group())
```

# Extracting the Email

```python
print(data[0:300])
```

```python
pattern = "\w+@\w+\.\w{3}"
ans = re.finditer(pattern,data)
for i, email in enumerate(ans):
#     print(email.start(),email.end(),email.group())
    print(email.group())
    
```

# Extracting the Names

```python
print(data[0:300])
```

```python
pattern = "[A-Z][a-z]*\s[A-Z][a-z]{2}"
ans  = re.finditer(pattern,data)

for i, name in enumerate(ans):
    print(name.group())
```
