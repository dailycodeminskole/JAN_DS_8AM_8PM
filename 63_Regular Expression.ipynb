```python
s = 'charlesharris@bogusemail.com'
s.split("@")
```

```python
def mask_email(s):
    if "@" in s:
        name , domain = s.split("@")
#         return name,domain
        return f'{name[0]}*******{name[-1]}@{domain}'
```

```python
mask_email('charlesharris@bogusemail.com')
```

```python
mask_email('abc.gmail.com')
```

```python
mask_email('r@gmail.com')
```

```python
import re
```

```python
def is_vemail(s):
    email_pattern = "^\w+([\-]?\w+) *@\w+([\-]?\w+)*(\.\w{2,3})+$"
    res = re.search(email_pattern,s)
    if res:
        return True
    else:
        return False
```

```python
is_vemail('abc.gmail.com')
```

```python
is_vemail('r@gmail.com')
```

```python
is_vemail('123@gmail.com')
```

```python
is_vemail('123@gmail@com')
```

```python
is_vemail('@123@gmail.com')
```
