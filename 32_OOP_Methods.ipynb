```python
class Account:
    counter = 0
    def __init__(self,openig_bal=0):
        Account.counter =  Account.counter +1
        self.id = Account.counter
        self.bal = openig_bal
        
    def deposit(self,amount):
        if amount > 0:
            self.bal = self.bal+amount
    def withdraw(self,amount):
        if amount > 0 and self.bal>= amount:
            self.bal = self.bal - amount
    def __str__(self):
        return "Ac Number : {} and Bal is Rs. : {}".format(self.id,self.bal)
```

```python
class Account:
    counter = 0
    def __init__(self,openig_bal=0):
        Account.counter =  Account.counter +1
        self.id = Account.counter
        self.bal = openig_bal
        
    def deposit(self,amount):
        if amount > 0:
            self.bal = self.bal+amount
    def withdraw(self,amount):
        if amount > 0 and self.bal>= amount:
            self.bal = self.bal - amount
    def __str__(self):
        return "Ac Number : {} and Bal is Rs. : {}".format(self.id,self.bal)
```

```python
class SavingsAccount(Account):
    pass
```

```python
class CurrentAccount(Account):
    pass
```

```python
s1 = SavingsAccount()
print(s1)
```

```python
c1 = CurrentAccount()
print(c1)
```
