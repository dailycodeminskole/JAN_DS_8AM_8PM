```python
# creating a function to  get the no of matches played
```

```python
def match_cities(city):
    mask2 = (data['city']==city)
    return data[mask2].shape[0]

match_cities('Hyderabad')
match_cities('Pune')
```

```python
mask3 = (data['team1'] == 'Delhi Daredevils')
```
