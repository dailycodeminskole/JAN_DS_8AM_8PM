# more complex plots

```python
# plotting square function 
x_val = [-3,-2,-1,0,1,2,3]
y_val = [9,4,1,0,1,4,9]
plt.plot(x_val,y_val)
```

```python
# plotting square function 
x_val = np.array([-3,-2,-1,0,1,2,3])
y_val = x_val**2 #using numpy array property : elementwise
plt.plot(x_val,y_val)
```

```python
# plotting square function 
x_val = np.arange(-3,3,.1)
y_val = x_val**2 #using numpy array property : elementwise
plt.plot(x_val,y_val)
```

```python
# plotting square function 
x_val = np.arange(-3,3,.1)
y_val = x_val**2 #using numpy array property : elementwise
plt.xticks(np.arange(-3,3,0.3),rotation=90)
plt.plot(x_val,y_val)
```

```python
# plotting square function 
x_val = np.arange(-3,3,.1)
y_val = x_val**3 #using numpy array property : elementwise
plt.xticks(np.arange(-3,3,0.3),rotation=90)
plt.plot(x_val,y_val)
plt.savefig("Cube-fig.jpg")
```
