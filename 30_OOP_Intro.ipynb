```python
class Student:
    def __init__(self,name):
        print(id(self))
#         self.z = name
        self.name = name
        

s1 = Student('Ravi')
print(id(s1))
print(s1.name)
```

```python
class Student:
    def __init__(self,name,age,area):

        self.name = name
        self.sss = age
        self.qqq = area

s1 = Student('Ravi',24,"Pune")
print(s1.name)
print(s1.sss)
print(s1.qqq)
```

```python
class Student:
    def __init__(self,name,age,area):

        self.name = name
        self.age = age
        self.area = area

s1 = Student('Ravi',24,"Pune")
print(s1.name)
print(s1.age)
print(s1.area)
```

```python
s2 = Student('Sameer',58,"Nashik")
print(s2.name)
print(s2.age)
print(s2.area)
```
