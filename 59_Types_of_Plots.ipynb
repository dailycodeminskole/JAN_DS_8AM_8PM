```python
import matplotlib.pyplot as plt
import numpy as np
```

# BAR 

```python
# shop1
# fruit1 :  10
# fruit2 : 20
# fruit3  :  5
```

```python
plt.bar([1,2,3],[10,20,5])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=0.2)
plt.legend("Shpo1")
plt.show()
```

```python
# shop2
# fruit1 :  5
# fruit2 : 10
# fruit3  :  6
```

```python
plt.bar([1,2,3],[5,10,6],width=0.5)
plt.legend("shop2")
plt.show()
```

```python
# plotting both the charts in one canvas
```

```python
plt.bar([1,2,3],[10,20,5],width=0.8) #shop1 data
plt.bar([1,2,3],[5,10,6],width=0.2) #shop2 data
plt.legend(['shpo1','shop2'])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=0.2) #shop1 data
plt.bar([1.2,2.2,3.2],[5,10,6],width=0.2) #shop2 data
plt.legend(['shpo1','shop2'])
plt.show()
```
